**Füllrate der Spalten (df_col_drop)**

```python
def analyze_fill_status(df_col_drop):
    # Berechnung der Füllstatistiken
    total_rows = len(df_col_drop)
    fill_stats = pd.DataFrame({
        'Spalte': df_col_drop.columns,
        'Gefüllte_Zeilen': df_col_drop.count(),
        'Leere_Zeilen': df_col_drop.isnull().sum(),
        'Füllrate (%)': (df_col_drop.count() / total_rows * 100).round(2)
    })
    
    fill_stats = fill_stats.sort_values('Füllrate (%)', ascending=False)
    
    # Ausgabe der Statistiken
    print("Füllstatistiken für alle Spalten:")
    print(fill_stats)
    print("\n")
    
    # Visualisierung der Füllraten
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Füllrate (%)', y='Spalte', data=fill_stats)
    plt.title('Füllrate der Spalten')
    plt.xlabel('Füllrate (%)')
    plt.ylabel('Spalten')
    plt.tight_layout()
    plt.show()
    
    # Detaillierte Analyse für jede Spalte
    for column in df_col_drop.columns:
        filled_rows = df_col_drop[column].count()
        empty_rows = df_col_drop[column].isnull().sum()
        unique_values = df_col_drop[column].nunique()
        
        print(f"\nAnalyse für Spalte '{column}':")
        print(f"Gesamtanzahl der Zeilen: {total_rows}")
        print(f"Gefüllte Zeilen: {filled_rows}")
        print(f"Leere Zeilen: {empty_rows}")
        print(f"Füllrate: {filled_rows/total_rows*100:.2f}%")
        print(f"Anzahl einzigartiger Werte: {unique_values}")
        
        # Zeige grundlegende Statistiken für numerische Spalten
        if pd.api.types.is_numeric_dtype(df_col_drop[column]):
            print("Grundlegende Statistiken:")
            print(df_col_drop[column].describe())
        
        print("-" * 50)

# Analyse durchführen
analyze_fill_status(df_col_drop)
```

**Füllrate der Zeilen kontrollieren**

```python
def analyze_row_fill_status(data_cleaned):
    # Berechnung der Füllstatistiken für Zeilen
    total_columns = len(data_cleaned.columns)
    row_fill_stats = pd.DataFrame({
        'Zeile': range(len(data_cleaned)),
        'Gefüllte_Spalten': data_cleaned.notna().sum(axis=1),
        'Leere_Spalten': data_cleaned.isna().sum(axis=1)
    })
    row_fill_stats['Füllrate (%)'] = (row_fill_stats['Gefüllte_Spalten'] / total_columns * 100).round(2)
    
    # Sortieren nach Füllrate
    row_fill_stats = row_fill_stats.sort_values('Füllrate (%)', ascending=False)
    
    # Ausgabe der Statistiken
    print("Füllstatistiken für alle Zeilen:")
    print(row_fill_stats.head(10))  # Zeigt die ersten 10 Zeilen
    print("\n")
    
    # Visualisierung der Füllraten
    #plt.figure(figsize=(12, 6))
    #sns.histplot(data=row_fill_stats, x='Füllrate (%)', bins=20)
    #plt.title('Verteilung der Zeilenfüllraten')
    #plt.xlabel('Füllrate (%)')
    #plt.ylabel('Anzahl der Zeilen')
    #plt.tight_layout()
    #plt.show()
    
    # Zusammenfassung der Zeilenfüllraten
    print("Zusammenfassung der Zeilenfüllraten:")
    print(row_fill_stats['Füllrate (%)'].describe())
    print("\n")
    
    # Identifizierung von Zeilen mit niedrigen Füllraten
    low_fill_threshold = 50  # Schwellenwert für niedrige Füllrate (in %)
    low_fill_rows = row_fill_stats[row_fill_stats['Füllrate (%)'] < low_fill_threshold]
    
    print(f"Zeilen mit Füllrate unter {low_fill_threshold}%:")
    print(low_fill_rows)
    print("\n")
    
    # Detaillierte Analyse für ausgewählte Zeilen
    print("Detaillierte Analyse für die ersten 5 Zeilen mit der niedrigsten Füllrate:")
    for index in low_fill_rows.head().index:
        print(f"\nZeile {index}:")
        print(data_cleaned.iloc[index])
        print(f"Füllrate: {row_fill_stats.loc[index, 'Füllrate (%)']}%")
        print("-" * 50)

# Führen Sie die Analyse durch
analyze_row_fill_status(data_cleaned)
```

**FRAGE: Wie normalisiere ich die Daten, wenn diese Normalverteilt und Nicht-Normalverteilt sind?**
- getrennt normalisieren?


**Daten normalisieren**

Sowohl Z-Score-Normalisierung als auch Min-Max-Skalierung sind nützliche Techniken zur Datennormalisierung, aber ihre Eignung hängt vom spezifischen Anwendungsfall und den Eigenschaften der Daten ab. Hier ist ein Vergleich der beiden Methoden:

## Z-Score-Normalisierung

### Vorteile:
1. Bewahrt Ausreißer und extreme Werte
2. Nützlich für Daten, die annähernd normalverteilt sind
3. Resultiert in Daten mit Mittelwert 0 und Standardabweichung 1

### Nachteile:
1. Nicht geeignet für Daten, die nicht normalverteilt sind
2. Kann zu Problemen führen, wenn die Standardabweichung sehr klein ist

## Min-Max-Skalierung

### Vorteile:
1. Skaliert Daten in einen festen Bereich (typischerweise [0, 1])
2. Bewahrt die ursprüngliche Verteilung der Daten
3. Nützlich für Daten, die nicht normalverteilt sind

### Nachteile:
1. Anfällig für Ausreißer, die die Skalierung stark beeinflussen können
2. Kann Informationen über die ursprüngliche Verteilung verlieren

## Wann welche Methode verwenden?

1. **Z-Score-Normalisierung** ist besser, wenn:
   - Die Daten annähernd normalverteilt sind
   - Ausreißer wichtige Informationen enthalten
   - Sie Merkmale mit unterschiedlichen Mittelwerten und Standardabweichungen vergleichen möchten

2. **Min-Max-Skalierung** ist besser, wenn:
   - Sie Daten in einen bestimmten Bereich bringen müssen (z.B. für neuronale Netze)
   - Die Verteilung nicht normalverteilt ist
   - Sie die relative Verteilung der Daten beibehalten möchten

## Fazit

Für maschinelles Lernen kann die Z-Score-Normalisierung vorteilhaft sein, da sie hilft, dass Modelle schneller konvergieren und besser mit Merkmalen unterschiedlicher Skalen umgehen können. Allerdings kann die Min-Max-Skalierung in Situationen, wo eine feste Skalierung erforderlich ist, wie bei bestimmten neuronalen Netzarchitekturen, die bessere Wahl sein.

**Mischung aus Z-Score (normalverteilte Daten) und Min-Max-Skalierung (Nicht Normalverteilte Daten) nutzen:**

```python
from scipy import stats
#Z-Score für normalverteilte Daten
def z_score_normalize(column):
    return stats.zscore(column)


#Min-Max-Skalierung für Nicht Normalverteilte Daten
from sklearn.preprocessing import MinMaxScaler

def min_max_scale(column):
    scaler = MinMaxScaler()
    return scaler.fit_transform(column.values.reshape(-1, 1)).flatten()


df_normalized = pd.DataFrame(df_cleaned)
for column in df_cleaned.select_dtypes(include='number').columns:
    if stats.normaltest(df_cleaned[column])[1] > 0.05:  # p-Wert > 0.05 deutet auf Normalverteilung hin
        df_normalized[column] = z_score_normalize(df_cleaned[column])
    else:
        # Wählen Sie eine der nicht-normalverteilten Methoden
        df_normalized[column] = min_max_scale(df_cleaned[column])
```

```python
# Beispieldaten
x = df_grouped_cont['total_cases']
y = df_grouped_cont['total_deaths']

# Berechnung der Korrelation
correlation = np.corrcoef(x, y)[0, 1]

print(f"Die Korrelation zwischen total_cases und total_deaths ist: {correlation}")
```

- Farben zeigen die Stärke und Richtung der Korrelationen an:
  
  - Rote Farben zeigen positive Korrelationen
  - Blaue Farben zeigen negative Korrelationen
  - Hellere Farben zeigen schwächere Korrelationen
  - Dunklere Farben zeigen stärkere Korrelationen
  
- Korrelationskoefizient:
  - 0,0 < 0,1	kein Zusammenhang
  - 0,1 < 0,3	geringer Zusammenhang
  - 0,3 < 0,5	mittlerer Zusammenhang
  - 0,5 < 0,7	hoher Zusammenhang
  - 0,7 < 1	sehr hoher Zusammenhang

- Der Wert +1 bedeutet, dass ein gänzlich positiver linearer Zusammenhang vorliegt (je mehr, desto mehr).
- Der Wert -1 gibt an, dass ein gänzlich negativer linearer Zusammenhang besteht (je mehr, desto weniger).
- Bei einem Wert von 0 liegt kein linearer Zusammenhang vor, die Variablen korrelieren nicht miteinander.

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objs as go
import pandas as pd

# Daten einlesen
df = pd.read_csv('df_grouped_cont.csv')

# Impfquote berechnen
df['impfquote'] = df['people_fully_vaccinated'] / df['population'] * 100

# Dash-App initialisieren
app = dash.Dash(__name__)

# Layout des Dashboards
app.layout = html.Div([
    html.H1("COVID-19 Dashboard nach Kontinenten"),
    
    dcc.Graph(id='correlation-heatmap'),
    
    dcc.Dropdown(
        id='barplot-dropdown',
        options=[
            {'label': 'Stringenz-Index und Impfquote', 'value': 'stringency-vaccination'},
            {'label': 'Testraten und Positivraten', 'value': 'test-positive'},
            {'label': 'Übersicht aller Metriken', 'value': 'overview'}
        ],
        value='stringency-vaccination',
        style={'width': '50%'}
    ),
    
    dcc.Graph(id='barplot')
])

# Callback für die Korrelationsheatmap
@app.callback(
    Output('correlation-heatmap', 'figure'),
    Input('correlation-heatmap', 'id')
)
def update_correlation_heatmap(_):
    corr_cols = ['total_cases', 'total_deaths', 'stringency_index', 'impfquote', 'positive_rate', 'gdp_per_capita', 'life_expectancy', 'human_development_index']
    corr_matrix = df[corr_cols].corr()
    fig = go.Figure(data=go.Heatmap(
        z=corr_matrix.values,
        x=corr_cols,
        y=corr_cols,
        colorscale='RdBu',
        zmin=-1, zmax=1,
        colorbar=dict(title='Korrelation')
    ))
    fig.update_layout(title='Korrelationsmatrix')
    return fig

# Callback für die Barplots
@app.callback(
    Output('barplot', 'figure'),
    Input('barplot-dropdown', 'value')
)
def update_barplot(selected_plot):
    if selected_plot == 'stringency-vaccination':
        fig = go.Figure()
        fig.add_trace(go.Bar(x=df['continent'], y=df['stringency_index']*100, name='Stringenz-Index', text=df['stringency_index']*100, textposition='outside'))
        fig.add_trace(go.Bar(x=df['continent'], y=df['impfquote'], name='Impfquote', text=df['impfquote'], textposition='outside'))
        fig.update_layout(title='Vergleich von Stringenz-Index und Impfquote nach Kontinent',
                          xaxis_title='Kontinent', yaxis_title='Prozent',
                          barmode='group', yaxis_range=[0, 100])
    elif selected_plot == 'test-positive':
        fig = go.Figure()
        fig.add_trace(go.Bar(x=df['continent'], y=df['new_tests']*100, name='Testrate', text=df['new_tests']*100, textposition='outside'))
        fig.add_trace(go.Bar(x=df['continent'], y=df['positive_rate']*100, name='Positivrate', text=df['positive_rate']*100, textposition='outside'))
        fig.update_layout(title='Testraten und Positivraten nach Kontinent',
                          xaxis_title='Kontinent', yaxis_title='Prozent',
                          barmode='group')
    else:  # overview
        fig = go.Figure()
        fig.add_trace(go.Bar(x=df['continent'], y=df['stringency_index']*100, name='Stringenz-Index', text=df['stringency_index']*100, textposition='outside'))
        fig.add_trace(go.Bar(x=df['continent'], y=df['impfquote'], name='Impfquote', text=df['impfquote'], textposition='outside'))
        fig.add_trace(go.Bar(x=df['continent'], y=df['total_cases']*100, name='COVID-19 Gesamtfälle', text=df['total_cases']*100, textposition='outside'))
        fig.add_trace(go.Bar(x=df['continent'], y=df['total_deaths']*100, name='COVID-19 Todesfälle', text=df['total_deaths']*100, textposition='outside'))
        fig.update_layout(title='Übersicht von Stringenz-Index, Impfquote, COVID-19 Gesamtfälle und Todesfälle nach Kontinent',
                          xaxis_title='Kontinent', yaxis_title='Prozent',
                          barmode='group')
    
    fig.update_traces(texttemplate='%{text:.2f}%', textposition='outside')
    fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
    return fig

# App starten
if __name__ == '__main__':
    app.run_server(debug=True)
```
